```python
gh_turnover = pd.read_csv("gh_turnover_four_projects.csv")
gh_turnover.head
```

### Rename Columns

```python
gh_turnover = gh_turnover.rename(columns={
    'domain': 'project_domain',
    'language': 'project_language',
    'windows': 'project_age',
    'window_idx': 'quarter',
    'num_team': 'team_size',
    'num_commits': 'project_commits',
    'blau_gender': 'gender_blau',
    'Gini_gh_ten': 'gh_tenure_gini',
    'Core1': 'core_dev',
    'commits': 'user_commits',
    'propCommits': 'commit_proportion',
    'github_tenure': 'gh_tenure',
    'leavesNextQ': 'leaves_next_q'
})
gh_turnover
```

### Data Wrangling

```python
gh_turnover['user_commits'].mean() # Mean of user_commits
```

```python
gh_turnover['user_commits'].median() # Median of user_commits
```

```python
turnover_new = gh_turnover.filter(regex='(^p)|g')  # starts with 'p' OR contains 'g'
turnover_new  # view selected column names
```

```python
turnover_new[turnover_new['project_commits'] > 9000] # commits>9000
```

```python
turnover_newer = (
    gh_turnover
    .filter(regex='(^p)|g')  
    .loc[gh_turnover['project_commits'] > 9000]
)
turnover_newer
```

```python
gh_filtered = (
    gh_turnover
    .filter(regex='(e$)|s')  # ends in e or has an s
    .loc[gh_turnover['gh_tenure'] >= 1991]  # filter rows on original df
)
gh_filtered
```

```python
gh_filtered['user_commits'].mean()
```

```python
gh_filtered['user_commits'].median()
```

The mean and the median are much lower than they were originally. They are also the same value. This is likely because ther are only two values in this data fram that have gh_tenure greater than 1991 this will cause the mean and the median to be the same value.
